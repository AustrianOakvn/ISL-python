```python
# Load data
train = pd.read_csv('../input/train.csv', index_col=0)
test = pd.read_csv('../input/test.csv', index_col=0)


# Concatenate training and test data into single dataframe for processing
df_all = pd.concat([train, test], axis=0, keys=['train', 'test'], names=['dataset'])
idx = pd.IndexSlice

# Convenience filter functions
def train(df): 
    """return only rows from training dataset"""
    t = df.loc[idx['train',:], :]
    return t

def test(df): 
    """return only rows from test dataset"""
    t = df.loc[idx['test',:], :]
    return t
```

# Functions

First some useful functions...

```python
def drop_columns(df, columns:list):
    """Remove column from datafram"""
    for col in columns:
        try:
            df.drop([col], axis=1, inplace=True)
        except KeyError as e:
            print('KeyError: {}'.format(e))

            
def fill_zero(df, columns:list):
    """Fill nan values with zeroes"""
    df[columns] = df[columns].fillna(0)
    return


def fill_mean(df, columns:list):
    """
    Fill nan values with mean of corresponding dataframe column
    Note: Mutates datafram in place"""
    for i in columns:
        df[i].fillna(df[i].mean(), inplace=True)
    return


def plot_correlation(df, constant:str, title=None):
    """Plot correlation between variables and constant"""
    variables = df.columns.drop(constant)
    data = pd.concat([df[constant], df[variables]], axis=1)
    corrmat = data.corr().sort_values(by=[constant], ascending=True)
    #plt.rcParams['figure.figsize'] = (8.0, 10.0)
    corrmat[constant].plot(kind='barh', color='b')
    plt.title(title)
    return


def plot_grid(df, constant):
    """ Plot all variables against constant in grid """
    variables = df.columns.drop(constant)
    for i in range(0, len(variables), 6):
        g = sns.PairGrid(df, y_vars=[constant], x_vars=variables[i:i+6], height=4)
        g.map(sns.regplot)
        g.set(ylim=(10, 15));
    return


def plot_hist(df):
    """histogram plot"""
    sns.distplot(df, fit=stats.norm);
    return


def plot_variance(df, title:str):
    rm_units(df).var().sort_values(ascending=True).plot(kind='barh', color='b');
    plt.title(title);
    plt.xlabel('Variance');
    return

def rm_units(df):
    """Remove units"""
    return (df-df.min())/(df.max() - df.min())


def rmse(y, y_predicted):
    """Measure Root-Mean-Square-Error (RMSE)"""
    return (mean_squared_error(y, y_predicted))**0.5


def scale(df):
    """Andrew ng's normalisation to remove units
    Xi = Xi - mu(Xi) / s(Xi)
    mu -> mean
    s -> standard deviation"""
    return (df-df.mean())/(df.std())
```

# Model validation

What approach should we take to validate our models?

Holdout is a simple approach where we 'hold back' some portion of our dataset for validation.

```python
# Hold-out validation
# ------------------------------------------------

# Plot
def plot_reg(x_train, y_train, x_test, y_test, y_predict):
    plt.scatter(x_train, y_train, color='b', alpha=0.3) # blue - training set
    plt.scatter(x_test, y_test, color='r', alpha=0.3) # red - evlauation set
    plt.plot(x_test, y_predict)
    plt.legend(('Model fit', 'Training set', 'Test set'), scatterpoints=1, frameon=False, labelspacing=1)
    plt.show();

def holdout_val(model,X, y, train_size, plot=False):
    # split the data with 50% in each set
    X1, X2, y1, y2 = train_test_split(X, y, random_state=0, train_size=train_size)
    # Predict
    y2_predict = model.fit(X1, y1).predict(X2)
    if plot:
        plot_reg(X1, y1, X2, y2, y2_predict)
    return rmse(y2, y2_predict)

# Load data
x = train(df_real)['GrLivArea'][:, np.newaxis]
y = np.log(train(df_all)['SalePrice']) 

# Load model
model = LinearRegression(fit_intercept=True)

# Evaluate
error = holdout_val(model, x, y, 0.7, plot=True)
print('Holdout validation (70/30 split)' + '\nRMSE: ' + str(error))
```

Here we perform univariate linear regression on the GrLivArea feature. We use holdout validation to hold back 30% of our training set to evaluate our model.

There are two potential drawbacks of this approach:
1. if we choose a test set that is too small then the RMSE score can be skewed, if there's a large proportion of anomolous datapoints in the test set
2. if we choose a test set that is too large our training set becomes small limiting our ability to inspect the benefits of more complex models that might only show benefits for a larger training set.

We can mitigate this by using cross-validation to perform multiple trials with different populations in the test set for each trial. We can see the importance of this by comparing validation approaches for multivariate linear regression with all of our features.


```python
from sklearn.model_selection import cross_val_score

def cross_val_2(model, X, y):
    """Two-way cross validation
    https://jakevdp.github.io/PythonDataScienceHandbook/05.03-hyperparameters-and-model-validation.html
    """
    # Split the data with 50% in each set
    X1, X2, y1, y2 = train_test_split(X, y, random_state=0, train_size=0.5)
    # Predict
    y1_predict = model.fit(X2, y2).predict(X1)
    y2_predict = model.fit(X1, y1).predict(X2)
    # Evaluate
    errors = np.array([rmse(y1, y1_predict), rmse(y2, y2_predict)])
    mean_error  = np.sum(errors)/2
    return mean_error, errors

def cross_val_n(model, X, y, trials):
    """Cross validation n-way
    Sklearns built in cross_val_score doesn't support root mean squared error scoring
    So here we convert the negative mean squared error... yes I know negative mean squared error
    makes no sense!... looks like it is an sklearn quirk, they decided to flip the sign for some reason!"""
    errors = (-1 * cross_val_score(model, X, y, cv=trials, scoring='neg_mean_squared_error'))**0.5
    error_avg = np.sum(errors)/trials
    return error_avg, errors


# Load data
X = pd.concat([train(df_real), train(df_count), train(df_ordinal), train(df_categorical)], axis=1)
y = np.log(train(df_all)['SalePrice']) 

# Load model
model = LinearRegression(fit_intercept=True)

# Compare validation approaches
holdout = holdout_val(model, X, y, 0.5)
print('Holdout validation (50/50 split)' + '\nRMSE: ' + str(holdout))
cross2_avg, cross2_errs = cross_val_2(model, X, y)
print('\n2-way cross validation (50/50 split)' + '\nRMSE avg.: ' + str(cross2_avg)+ '\nRMSE per trial: ' + str(cross2_errs))
cross_n_avg, cross_n_errs = cross_val_n(model, X, y, 10)
print('\n10-way cross validation (90/10 split)' + '\nRMSE avg.: ' + str(cross_n_avg) + '\nRMSE per trial: ' + str(cross_n_errs))
```

The observed error of our model decreases as we introduce cross validation and increase the number of trials. Why?

- the model is same for all tests so our model isn't getting any better, which do we trust?
- comparing the results for individual trials in cross validation we see high variation, suggesting that how the data is split has a significant effect on validation results
- cross validation with a large number of trials allows our model to train on a large proportion of the dataset, this would allow more complex models to perform better but we can 't see that result here explicitly

10-way cross validation allows us to remove some of the noise caused by anomalies in our dataset and might allow us to better observe the benefits of more complex models. We'll use it to test our models from here on in.

# Feature performance

Let's unpack the above result to see if we can see how much each of our feature types is contributing.

```python
def compare_feature_types(model, Xt, y):
    def combination_matrix(items:list):
        combos = np.array(np.meshgrid([items[0], None], [items[1], None], [items[2], None], [items[3], None])).T.reshape(-1, 4)
        return combos[:-1] # exclude item where all none
    
    # Parse inputs
    df_real        = Xt[0]
    df_count       = Xt[1]
    df_ordinal     = Xt[2]
    df_categorical = Xt[3]

    # Create matrix of all label combinations
    label_names = ['df_real', 'df_count', 'df_ordinal', 'df_categorical']
    labels = combination_matrix(label_names)
    labels = pd.DataFrame(labels).fillna('-')
    
    # Create matrix of feature type combination
    data_combos = combination_matrix([df_real, df_count, df_ordinal, df_categorical])
    
    # Score every combination of feature types
    scores = []
    for i in range(len(data_combos)):
        X = pd.concat(data_combos[i], axis=1)
        cross_n, _ = cross_val_n(model, X, y, 10)
        scores.append(cross_n)
    
    # Plot results table
    scores = pd.DataFrame(scores, columns=['RMSE'])
    return pd.concat([labels, scores], axis=1).sort_values(by=['RMSE'])

# Compare features
Xt = [train(df_real), train(df_count), train(df_ordinal), train(df_categorical)]
y = np.log(train(df_all)['SalePrice'])
model = LinearRegression(fit_intercept=True)
compare_feature_types(model, Xt, y)
```

Here we can see the RSME score for every combination of feature types. The lowest error of 1.469805e-01 is achieved by providing all features to our basic model. The highest error results from using only the categorical features. 

We can see that the scores for feature types in isolation are as follows:

| datatype | RMSE|
| --- | --- |
| df_real | 1.867763e-01 |
| df_ordinal | 1.912816e-01 |
| df_count | 2.111544e-01 |
| df_categorical | 1.615918e+10 |
  
The categorical perform very poorly in isolation but still seem to provide a marginal improvement when provided along with *all* other feature types.

Interestingly use of categorical features results in a very poor score without real features, or without ordinal features, suggesting some kind of interaction. **How might categorical features only add information when combined with real or ordinal features?**

Let's park that question for now, but keep in mind that our features are interacting with one another in interesting ways as we try to increase the strength of our features.




# Feature normality

Let's start by seeing if we can increase the strength of our real features. We can see from the plots in feature engineering that some of these features might have non-linear relationships with SalePrice, that will limit our linear model's ability to fit the data properly. We can measure the linearity of these relationships using the `scipy.stats.skew` function.

```python
from scipy.stats import skew

# Measure skews
skews = df_real.skew() # compute skewness

# Log(feature + 1) transform to reduce skew
# log1p incase of zeros
df_real_norm = np.log1p(df_real)

# CV
Xt = [train(df_real_norm), train(df_count), train(df_ordinal), train(df_categorical)]
y = np.log(train(df_all)['SalePrice'])
model = LinearRegression(fit_intercept=True)
compare_feature_types(model, Xt, y)
```

Our best result is now  1.335950e-01 rmse, about a 9% improvement on our previous score. 

The strongest feature set now excludes categorical features. 

Suprisingly the use of all features, which scored 1.469805e-01 before normalisation, now scores 1.304992e+10.  **What's going on here?**

```python
# Measure skew of transformed data and compare
skews_norm = df_real_norm.skew()
skews.name = 'Pre Transform'
skews_norm.name = 'Post Transform'

# Results
reduction = (skews_norm.abs()-skews)
reduction.name = 'Diff'
skew_results = pd.concat([skews.abs(), skews_norm.abs(), reduction], axis=1)

# Plot
ax = skew_results['Diff'].plot(kind='barh', figsize=(10, 10))
ax.set_xlabel('ΔSkew')
ax.set_ylabel('Features');
```

We can see that we've reduced skew for some feature but not for others. Let's only apply our transform if it is effective.

```python
do_norm = skew_results[skew_results['Diff'] < 0].index
dont_norm = skew_results[skew_results['Diff'] >= 0].index

normed = np.log1p(df_real[do_norm])
unnormed = df_real[dont_norm]
df_real_norm_selective  = pd.concat([normed, unnormed], axis=1)
```

Ok, lets look at our results again using the normalised real features.

```python
# CV
Xt = [train(df_real_norm_selective), train(df_count), train(df_ordinal), train(df_categorical)]
y = np.log(train(df_all)['SalePrice'])
model = LinearRegression(fit_intercept=True)
compare_feature_types(model, Xt, y)
```

The error is higher than when we just apply our transform to all features. What if we naively transform the count or ordinal features?

```python
df_count_norm = np.log1p(df_count)

# CV
Xt = [train(df_real_norm_selective), train(df_count_norm), train(df_ordinal), train(df_categorical)]
y = np.log(train(df_all)['SalePrice'])
model = LinearRegression(fit_intercept=True)
compare_feature_types(model, Xt, y)
```

```python
df_ordinal_norm = np.log1p(df_ordinal)

# CV
Xt = [train(df_real_norm_selective), train(df_count), train(df_ordinal_norm), train(df_categorical)]
y = np.log(train(df_all)['SalePrice'])
model = LinearRegression(fit_intercept=True)
compare_feature_types(model, Xt, y)
```

Normalising count or ordinal features in addition to real features yields a higher error than normalising real features only. 

**Conclusion**

- Naive normalisation of real features only, yields the lowest error score. 
- Selectively normalising features for which skew is reduced by the transform yields no beneftit. 
- Additional normalisation of count and ordinal features does not improve results.

# Outliers

Can we improve our result by removing anomalies from my training set? The hypothesis is that removing anomalies would allow a linear model to better fit the majority of data, at the cost of a worst fit for anomalies. My assumption here was that the error in fitting anomalies would be outweighed by better fit for majority of datapoints.

## Isolation forest

Isolation Forest is a novel technique for identifying anomlies in a high-dimensional dataset. It has the adavantage, compared to say Dbscan, that it is not as sensitive to the configuration of hyperparameters, and so it is relatively easy to implement.

Here I implement Isolation Forest to identify anomalies in our training dataset.

```python
from sklearn.ensemble import IsolationForest

rng = np.random.RandomState(42)

# Load data
X_train = pd.concat([train(df_real_norm), train(df_count), train(df_ordinal), train(df_categorical)], axis=1)

# Fit the model
clf = IsolationForest(max_samples=100, random_state=rng)
clf.fit(X_train)

# Average anomaly score of X of the base classifiers.
anomaly_score = pd.DataFrame(clf.decision_function(X_train), columns=['AnomalyScore'])
# For each observations, tells whether or not (+1 or -1) it should be considered as an inlier according to the fitted model.
is_inlier = pd.DataFrame(clf.predict(X_train), columns=['IsInlier'])

outlier_ids  = is_inlier[is_inlier['IsInlier'] < 0].index + 1

print('Outliers: ' + str(len(outlier_ids)))
```

```python
def drop_ids(df, ids):
    new_df = df.drop(index=ids, level=1, axis=0, inplace=False)
    return new_df
```

We can't use our existing cv function here, because anomalies will be removed from the validation set, and so any improvement measured could be the result of over-fitting.

Instead we need a new CV that allows us only to remove outliers from the training set, and not the hold-out set.

```python
# Load data
X = train(pd.concat([df_real_norm, df_count, df_ordinal], axis=1))
y = np.log(train(df_all)['SalePrice'])

# Split the data with 50% in each set
X1, X2, y1, y2 = train_test_split(X, y, random_state=0, train_size=0.7)

# Drop outliers from training set
X3 = drop_ids(X1, outlier_ids)
y3 = drop_ids(y1, outlier_ids)

# Load model
model = LinearRegression(fit_intercept=True)

# Predict
y2_predict = model.fit(X1, y1).predict(X2)
# Evaluate
print(rmse(y2, y2_predict))

# Predict
y2_predict = model.fit(X3, y3).predict(X2)
# Evaluate
print('drop outliers:')
print(rmse(y2, y2_predict))
```
